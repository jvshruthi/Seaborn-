# **SEABORN LIBRARY IN PYTHON**

Seaborn is a library mostly used for statistical plotting in Python. It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive.
The following plots are used efficiently
1.distplot
2.jointplot
3.pairplot
4.catplot
5.barplot
6.boxplot
7.stripplot
8.swarmplot
9.heatmap
10.lmplot

```python
import pandas as pd
data=pd.read_csv("../input/automobile-dataset/Automobile_data.csv")
```

# ****1.Plotting univariate distributions****

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, axes = plt.subplots(3, 3)
fig=figsize=(8,8)

#create boxplot in each subplot
sns.distplot(data['engine-size'], ax=axes[0,0])
sns.distplot(data['horsepower'], ax=axes[0,1])
sns.distplot(data['peak-rpm'], ax=axes[0,2])
sns.distplot(data['highway-mpg'], ax=axes[1,0])
sns.distplot(data['wheel-base'], ax=axes[1,1])
sns.distplot(data['city-mpg'], ax=axes[1,2])
sns.distplot(data['compression-ratio'], ax=axes[2,0])
sns.distplot(data['symboling'], ax=axes[2,1])
sns.distplot(data['height'], ax=axes[2,2])
```

```python
import seaborn as sns
sns.distplot(data['engine-size'])
```

**More engines sizes are distributed between 100-150.Thus the common used machine size is 100**

```python
sns.distplot(data['bore'])
```

**The maximumum bore value is between 3-4**

```python
sns.distplot(data['peak-rpm'])
```

**The peak rpm is distributed between 4000 - 6000 and the highest common value is around 5000 rpm**

```python
data['horsepower']=data['horsepower'].astype(dtype="float")
data['peak-rpm']=data['peak-rpm'].astype(dtype="float")
data['stroke']=data['stroke'].astype(dtype="float")
data['price']=data['price'].astype(dtype="float")
```

# **Plotting bivariate distributions**

```python
sns.jointplot(data['engine-size'],data['horsepower'])
```

**In this graph we can infer 3 things:**
1. when the engine size increases also the horsepower value increases
2. More number of engines produced with engine size 100 
3. More engines are produced with horsepower 50-100  

```python
sns.jointplot(data['wheel-base'],data['peak-rpm'])
```

**The most prominent observation is the peak rpm remains constant irrespective of any wheel base**

# **2.Visualising pairwise plots**

```python
sns.pairplot(data[['engine-size','city-mpg','compression-ratio','width']])
```

**We can conclude more observations comparing each variables**
1.For various width values the compression ratio and the peak-rpm remains nearly constant with some higher compression ratios and linearly decreases with city mpg
2.Compression ratio are mostly 10 and 20 and 10 being the most commonly used.
3.City mpg increases with decreases in size.

```python
sns.barplot(data['num-of-doors'],data['length'],hue=data['engine-type'])
```

```python
sns.boxplot(data['aspiration'],data['peak-rpm'],hue=data['fuel-type'])
```

```python
sns.countplot(data['drive-wheels'],hue=data['fuel-type'])
```

```python
sns.lmplot(y="horsepower",x="highway-mpg",hue="fuel-type",data=data)
```

**horsepower drastic decreases with increases with horsepower in gas compared with diesel.**

```python
sns.catplot(x="fuel-type",y="price",hue="num-of-doors",col="drive-wheels",data=data,kind="bar")
```

**Diesel vehicle is costing high compared with gas with rwd vehicles **

```python
sns.catplot(x="fuel-type",y="price",hue="fuel-system",col="num-of-doors",data=data,kind="bar")
```

*Vehicles with two doors are most common and pricing 10000 to 20000*

```python
sns.catplot(x="fuel-type",y="price",hue="num-of-doors",col="aspiration",data=data,kind="violin")
```

# **4.Correlations with all the variable**

```python
data.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr(),vmin=-1,
            cmap='coolwarm',
            annot=True);
```

The factors that affects the pricing includes:
1.Engine size and curb-weight affects the price about 80%
2.Wheel base,length,horsepower and width affects the price by 60-70%
3.City mpg and highway mpg ranges between 60-70% affects negatively with price
4.All other factors have least effect but stroke influences engine size about 30%
